```python
brca1DNA = []
for i in range(0, len(brca1Variants), 3):
    hgvs = str(brca1Variants["nthgvs"][i])
    if(hgvs != "nan"):
        count = 0
        p = str(brca1Variants["pos"][i])
        for j in range(0, len(p)):
            if(p[j] == '-' or p[j] == '+'):
                count = count + 1
        if(count == 0):
            brca1DNA.append(brca1Variants["nthgvs"][i][-3])
            

print(brca1DNA)
len(brca1DNA)
```

```python
a = 1
b = 1
c = 0
d = 0
position = 1
for i in range(0, len(brca1Variants)):
    hgvs = str(brca1Variants["nthgvs"][i])
    if(hgvs != "nan"):
        count = 0
        p = str(brca1Variants["pos"][i])
        for j in range(0, len(brca1Variants["pos"][i])):
            if(p[j] == '-' or p[j] == '+'):
                count = count + 1
        if(count == 0):
            c = c + 1
            if(c%9 == 1):
                d = d + 1
            
            pos = int(hgvs[2:-3])
            prev = hgvs[-3]
            prec = hgvs[-1]
            pPrior = str(brca1Variants["protein_prior"][i])
            aPrior = str(brca1Variants["applicable_prior"][i])
            hg19 = str(int(brca1Variants["hg19pos"][i]))
            
            
                
            if(a == 1):
                prevCodon = str(brca1DNA[position-1]) + str(brca1DNA[position]) + str(brca1DNA[position+1])
                prevCodon = str(codonTable[prevCodon])
            elif(a == 2):
                prevCodon = str(brca1DNA[position-2]) + str(brca1DNA[position-1]) + str(brca1DNA[position])
                prevCodon = str(codonTable[prevCodon])
            elif(a == 3):
                prevCodon = str(brca1DNA[position-3]) + str(brca1DNA[position-2]) + str(brca1DNA[position-1])
                prevCodon = str(codonTable[prevCodon])
                
            if(a == 1):
                precCodon = str(prec) + str(brca1DNA[position]) + str(brca1DNA[position+1])
                precCodon = str(codonTable[precCodon])
            elif(a == 2):
                precCodon = str(brca1DNA[position-2]) + str(prec) + str(brca1DNA[position])
                precCodon = str(codonTable[precCodon])
            elif(a == 3):
                precCodon = str(brca1DNA[position-3]) + str(brca1DNA[position-2]) + str(prec)
                precCodon = str(codonTable[precCodon])
            
            print(str(c) + "," +  str(position) + "," +  str(d) + "," +  str(b) + "," +  str(a) + "," + "NP_009225," + prevCodon + str(math.ceil((pos)/3))+ precCodon + ",c." + str(pos) + prev + ">" + prec + "," +  pPrior + "," + aPrior + "," + hg19)
            
            if(b%3 == 0):
                a = a + 1
                position = position + 1
            b = b + 1
            if(b == 4):
                b = 1
            if(c%9 == 0):
                a = 1
            
```
